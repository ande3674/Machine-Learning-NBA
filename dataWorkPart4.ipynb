```python
import requests
import pandas as pd
import time
```

```python
# Create empty dataframe
col_names = ['playerID', 'gamesPlayed', 'season', 'min', 
             'fgm', 'fga', 'fg3m', 'fg3a', 'ftm', 'fta', 
             'oreb', 'dreb', 'reb', 'ast', 'stl', 'blk', 'turnover', 'pf', 
             'pts', 'fg_pct', 'fg3_pct', 'ft_pct']
players_stats_df = pd.DataFrame(columns=col_names)
```

```python
col_names2 = ['playerID', 'firstName', 'lastName', 'teamID', 'teamAbbrev']
players_df = pd.DataFrame(columns=col_names2)
```

## PLAYERS

```python
#url_players = 'https://www.balldontlie.io/api/v1/players?per_page=100'
url_all_games_2019 = 'https://www.balldontlie.io/api/v1/stats?seasons[]=2019&per_page=100&page=1'

player_id_list_2019 = []

init_data = requests.get(url_all_games_2019).json()
total_pages = init_data['meta']['total_pages']
print(f'Total pages: {total_pages}')

for page in range(1, total_pages+1):
    url = f'https://www.balldontlie.io/api/v1/stats?seasons[]=2019&per_page=100&page={page}'
    data = requests.get(url).json()
    print(f'Page: {page}')
    
    for d in data['data']:
        player_id = d['player']['id']
        
        if player_id not in player_id_list_2019:
            player_id_list_2019.append(player_id)
     
    print(f'Length of player id list after page {page} completed: {len(player_id_list_2019)}')
    print('Sleeping 1 second...')
    time.sleep(1)
    
player_id_list_2019 
```

### Player Data

```python
# Get data for each player in the 2019 player list

for pid in player_id_list_2019:
    player_url = f'https://www.balldontlie.io/api/v1/players/{pid}'
    player_data = requests.get(player_url).json()
    
    # Get the data to fill a row in the df: ['playerID', 'firstName', 'lastName', 'teamID', 'teamAbbrev']
    players_df.loc[len(players_df.index)] = [ player_data['id'], player_data['first_name'], 
                                             player_data['last_name'], player_data['team']['id'], 
                                             player_data['team']['abbreviation'] ]
    
    print(f'Data complete for player {pid}. Sleeping...')
    time.sleep(1)
    

players_df.head(10)   
```

## SEASON AVERAGES

```python
# Gather 2019 season averages for each 2019 player
for pid in player_id_list_2019[521:]:
    avgs_url = f'https://www.balldontlie.io/api/v1/season_averages?player_ids[]={pid}'
    player_data = requests.get(avgs_url).json()['data'][0]
    
    # Get the data to fill a row in the df
    players_stats_df.loc[len(players_stats_df.index)] = [ player_data['player_id'], player_data['games_played'], 
                                                        player_data['season'], player_data['min'], player_data['fgm'], 
                                                        player_data['fga'], player_data['fg3m'], player_data['fg3a'], 
                                                        player_data['ftm'], player_data['fta'], player_data['oreb'], 
                                                        player_data['dreb'], player_data['reb'], player_data['ast'], 
                                                        player_data['stl'], player_data['blk'], player_data['turnover'],
                                                        player_data['pf'], player_data['pts'], player_data['fg_pct'], 
                                                        player_data['fg3_pct'], player_data['ft_pct'] ]
    
    print(f'Data complete for player {pid}. Sleeping...')
    time.sleep(1)
    
players_stats_df.head(10)
```
